```python
import pandas as pd
import numpy as np

GDELT_data = pd.read_csv('gdelt_events_with_market_impact.csv')
```

```python
GDELT_sliced = GDELT_data[
    (GDELT_data['event_year'] == 2015) & (GDELT_data['market_impact'] == 1)
].copy()
```

```python
GDELT_sliced.columns
```

```python
# 1. Sort the DataFrame by 'event_time'
# Use ascending=True for oldest to newest, or False for newest to oldest
GDELT_sorted = GDELT_sliced.sort_values(by='event_time', ascending=True)

# 2. Write the sorted data to the file
with open('gdelt_output.txt', 'w', encoding='utf-8') as f:
    for row in GDELT_sorted.itertuples():
        f.write(f"TIME: {row.event_time}\n")
        f.write(f"SUMMARY: {row.event_summary}\n")
        f.write("-" * 50 + "\n\n")

print("Sorted file 'gdelt_output.txt' has been created!")
```
